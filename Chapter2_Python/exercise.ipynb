```python
import numpy as np
import matplotlib.pyplot as plt

def e_function(exponents):
    result = []
    for exponent in exponents:
        result.append(np.exp(exponent))

    return result

a = 1
b = 5

values = range(1, 6, 1)
result = e_function(values)

plt.plot(values, result, color="blue")
plt.xlabel("x")
plt.ylabel("e(x)")
plt.show()
```
